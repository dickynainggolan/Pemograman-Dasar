```python
import pandas as pd
import numpy as np # Digunakan untuk mengganti nilai

# --- 1. Memuat Data ---
try:
    # Membaca file CSV yang telah diunggah
    df = pd.read_csv('mahasiswa (1).csv')
    print("‚úÖ Data berhasil dimuat. (5 baris pertama):")
    print(df.head())
except FileNotFoundError:
    print("‚ùå Error: Pastikan file 'mahasiswa (1).csv' ada di lokasi yang benar.")
    exit()

print("\n" + "="*50 + "\n")
```

```python
# --- 2. Melihat Struktur Data (Inspeksi) ---
print("üîç Informasi Data (Tipe Data & Jumlah Non-Null):")
df.info()

print("\nüîé Jumlah Nilai Hilang (NaN/Null) per Kolom:")
print(df.isnull().sum())

print("\n" + "="*50 + "\n")
```

```python
# --- 3. Membersihkan Data (Data Cleaning) ---

# Kolom 'umur'
print("üßπ Pembersihan Kolom 'umur'...")
# Mengganti nilai string non-angka ('twenty', '21years') dan nilai aneh (150, 120, -5) dengan NaN
df['umur'] = df['umur'].replace({'twenty': np.nan, '21years': np.nan}).astype(str)
df['umur'] = df['umur'].str.replace(r'[^0-9.]', '', regex=True)

# Konversi ke numerik, error='coerce' akan mengubah nilai yang tidak bisa dikonversi menjadi NaN
df['umur'] = pd.to_numeric(df['umur'], errors='coerce')

# Menentukan batas umur yang realistis (misalnya, 18 hingga 40)
# Nilai yang di luar rentang ini dianggap tidak valid dan diubah menjadi NaN
df.loc[((df['umur'] < 18) | (df['umur'] > 40)), 'umur'] = np.nan

# Mengisi (Impute) Nilai Hilang di 'umur' dengan median
median_umur = df['umur'].median()
df['umur'].fillna(median_umur, inplace=True)

# Kolom 'nilai_akhir'
print("üßπ Pembersihan Kolom 'nilai_akhir'...")
# Mengganti nilai string non-angka ('999', '-20', 'ninety') dengan NaN
df['nilai_akhir'] = df['nilai_akhir'].replace({'999': np.nan, '-20': np.nan, 'ninety': np.nan})

# Konversi ke numerik
df['nilai_akhir'] = pd.to_numeric(df['nilai_akhir'], errors='coerce')

# Mengisi (Impute) Nilai Hilang di 'nilai_akhir' dengan median
median_nilai = df['nilai_akhir'].median()
df['nilai_akhir'].fillna(median_nilai, inplace=True)


# Kolom 'kehadiran_pct'
print("üßπ Pembersihan Kolom 'kehadiran_pct'...")
# Menentukan batas realistis (misalnya 0 sampai 100)
df.loc[((df['kehadiran_pct'] < 0) | (df['kehadiran_pct'] > 100)), 'kehadiran_pct'] = np.nan

# Mengisi Nilai Hilang di 'kehadiran_pct' dengan rata-rata (mean)
mean_kehadiran = df['kehadiran_pct'].mean()
df['kehadiran_pct'].fillna(mean_kehadiran, inplace=True)


# Kolom 'jurusan'
print("üßπ Pembersihan Kolom 'jurusan' (Mengatasi Nilai Hilang/Kosong)...")
# Mengisi nilai yang hilang/kosong dengan mode (nilai yang paling sering muncul)
mode_jurusan = df['jurusan'].mode()[0]
df['jurusan'].fillna(mode_jurusan, inplace=True)
# Mengubah casing 'teknik' menjadi 'Teknik' (Standardisasi)
df['jurusan'] = df['jurusan'].replace({'teknik': 'Teknik'})
```

```python
# --- 4. Mengubah Tipe Data (Type Casting) ---
print("üîÑ Mengubah Tipe Data...")
# Mengubah 'umur' menjadi integer (jika tidak ada desimal)
df['umur'] = df['umur'].astype(int)
# Mengubah 'tanggal_daftar' menjadi tipe datetime
# 'coerce' akan mengubah tanggal yang tidak valid (seperti 'May 22 2024') menjadi NaT
df['tanggal_daftar'] = pd.to_datetime(df['tanggal_daftar'], errors='coerce')
```

```python
# --- 5. Membuat Kolom Baru (Feature Engineering Sederhana) ---
print("‚ú® Membuat Kolom Baru: 'status_kelulusan'...")
# Contoh sederhana: Lulus jika Nilai Akhir > 70
df['status_kelulusan'] = np.where(df['nilai_akhir'] > 70, 'LULUS', 'TIDAK LULUS')
```
